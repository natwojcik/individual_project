```python
#Now I'll create a new column and mebrge information from loan_status to assess whether the loan was late in repayment or not
# "Current", "Fully Paid", "In Grace Period" and "Does not meet the credit policy. Status:Fully Paid" - will be "on_time"
# "Charged Off", "Late (31-120 days)", "Late (16-30 days)", "Default", "Does not meet the credit policy. Status:Charged Off" - will be "late"
```

```python
loan["loan_status"].isnull().sum()
```

```python
loan.loc[((loan["loan_status"] == "Current") | 
          (loan["loan_status"] == "Fully Paid") | 
          (loan["loan_status"] == "In Grace Period") | 
          (loan["loan_status"] == "Does not meet the credit policy. Status:Fully Paid")), "is_late"] = "on_time"

loan.loc[~((loan["loan_status"] == "Current") | 
          (loan["loan_status"] == "Fully Paid") | 
          (loan["loan_status"] == "In Grace Period") | 
          (loan["loan_status"] == "Does not meet the credit policy. Status:Fully Paid")), "is_late"] = "late"
```

```python
loan["is_late"]
```

```python
is_late_per = loan["is_late"].value_counts()*100/len(loan)
print(is_late_per)
```

```python
is_late_per = is_late_per.round()
print(is_late_per)
```

```python
labels = ["On Time", "Late"]
palette_color = sns.color_palette("pastel")
plt.pie(is_late_per, colors=palette_color, labels=labels)
```
